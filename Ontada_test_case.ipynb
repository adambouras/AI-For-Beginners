Glancing at the table, I can identify as a drug codes **NDDF_GCN** (see figure above); these codes can be linked to NDC codes to get more information either by using RXnorm, which is open source database from NIH; or use commercial drug databases such as:
* First Databank (FDB),
* Micromedex, and
* Multum
* MediSpan
* Gold Standard

```python
#make the rxcui api call to retrieve NDC codes
from urllib.request import urlopen
import json
# build random request when send API request
import random
import time
def rxnorm_ndc_call(RXCUI_CODE):
  try:
    response_ndc =urlopen(f"https://rxnav.nlm.nih.gov/REST/rxcui/{str(RXCUI_CODE)}/ndcs.json")
  except:
    pass
  rxcui_ndc_json = json.loads(response_ndc.read())
  return rxcui_ndc_json
```

```python
#Mapp CGN codes to rxcui codes
def rxnormcall(GCN_CODE):
  #make api call to receive rxcui codes
  try:
    response_rxcui =urlopen(f"https://rxnav.nlm.nih.gov/REST/rxcui.json?idtype=GCN_SEQNO&id={str(GCN_CODE)}")
  except:
    pass

  rxcui_json = json.loads(response_rxcui.read())

  for k in rxcui_json:
  # print(json_file[k])
    for kk in rxcui_json[k]:
      for code in rxcui_json[k][kk]:
        print(code)
        try:
          response_ndc_json = rxnorm_ndc_call(code)
          ndc_list = response_ndc_json['ndcGroup']['ndcList']['ndc']
        except:
          pass

  return rxcui_json, ndc_list
```

```python
#convert NDDF_GCN into columns for further processing
for i, c in enumerate(treatment_df['NDDF_GCN']):
  gcn_list = c.split("|")
  # using set()
  # to remove duplicated
  # from list
  gcn_list = list(set(gcn_list))
  if i ==1:
    for code in gcn_list:
      secondstowait = random.randint(10, 40)
      time.sleep(secondstowait)
      rxcui_res, ndc_res  = rxnormcall(code)
      print(code, rxcui_res,  ndc_res)
```
